### View Sample of Training Images

```python
def plot_example_images(samples_per_expression):
    plt.figure(0, figsize=(12,20))
    index = 0

    for expression in os.listdir("train/"):
        samples = random.sample(os.listdir("train/" + expression), samples_per_expression)
        for i,sample in enumerate(samples,1):
            index += 1
            plt.subplot(len(os.listdir("train/")),samples_per_expression,index)
            img = load_img(os.path.join("train", expression, sample), target_size=(48,48))
            plt.imshow(img, cmap="gray")
            plt.title(expression + " " + str(i))

    plt.tight_layout()
    plt.show()
    return

plot_example_images(5)
```
